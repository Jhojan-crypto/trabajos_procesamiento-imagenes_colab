```python
import cv2
import numpy as np

refPt = []
cropping = False

def cortar(event, x, y, flags, params):
    global refPt, cropping
    if event == cv2.EVENT_LBUTTONDOWN:
        refPt = [(x, y)]
        cropping = True
    elif event == cv2.EVENT_LBUTTONUP:
        refPt.append((x, y))
        cropping = False
        cv2.rectangle(clonel, refPt[0], refPt[1], (0, 255, 0), 2)
        cv2.imshow('image', clonel)

image = cv2.imread('/content/drive/MyDrive/Procesamiento_QuispeLopez/Unidad 2/Tarea05/laboratorio 10 crop/images/3.jpg')

#if image is None:
    #print("Error: Image not loaded. Please check the image path.")
#else:
for i in range(2):
    x, y, z = np.shape(image)
    clonel = cv2.resize(image, (144, 354), interpolation=cv2.INTER_AREA)
    clone = clonel.copy()
    cv2.namedWindow('image')
    cv2.setMouseCallback('image', cortar)

    while True:
        cv2.imshow('image', clonel)
        key = cv2.waitKey(1) & 0xFF
        if key == ord("r"):
            clonel = clone.copy()
            print("recorte nuevamente la imagen")
        if key == ord("c"):
            print("imagen cortada, presione una tecla para continuar")
            break

    if len(refPt) == 2:
        roi = clone[refPt[0][1]:refPt[1][1], refPt[0][0]:refPt[1][0]]
        roil = cv2.resize(roi, (144, 384), interpolation=cv2.INTER_AREA)

        i = i + 25
        cv2.imwrite('wititi' + str(i) + '.jpg', roil)

        cv2.waitKey(0)
        print("Completado...")
        cv2.destroyAllWindows()
```
